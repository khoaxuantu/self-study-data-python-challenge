```python
from bs4 import BeautifulSoup
import requests as re_
import pandas as pd

url = 'https://coinmarketcap.com/'

crypto_context = re_.get(url)
crypto_table = BeautifulSoup(crypto_context.text, 'html.parser')
```

```python
#Get columns title:
title_source = crypto_table.find_all('th')[1:-2]

title_list = [item.text for item in title_source]
title_list.append('TimeStamp')

print(title_list)
```

```python
from datetime import datetime as dt_
# Find context:
body_source = crypto_table.find_all('tr')[1:10]

# Extract context:
all_context = []
for item in body_source:
    body_context = item.find_all('td')[1:-2]
    individual_row = []
    for ind, context_ in enumerate(body_context):
        if (ind == 1):
            values = context_.find_all('p')[1].text
        elif (ind == 7):
            values = context_.find_all('p')[0].text
        elif (ind == 6):
            values = context_.find_all('span')[1].text
        else:
            values = context_.text
        individual_row.append(values)
    individual_row.append(dt_.now())

    all_context.append(individual_row)

df_body = pd.DataFrame(all_context)
df_body.head()
```

```python
df_body.columns = title_list

df_body.index = df_body['#']
df_body = df_body.drop(columns=['#'])

df_body.head()
```

```python
# Extract to csv file

filename = './crypto_challenge.csv'
df_body.to_csv(filename)
```
