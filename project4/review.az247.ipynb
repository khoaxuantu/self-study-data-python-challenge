```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://coinmarketcap.com/"
page = requests.get(url)
soup = BeautifulSoup(page.text, "html")
table = soup.find("table", "sc-7b3ac367-3 etbcea cmc-table")

table_columns = table.find_all("th")
table_header = [header.text for header in table_columns]

df = pd.DataFrame(columns=table_header)
df

table_rows = table.find_all("tr")
for i in range(1, 10):
    individual_row = [i.text for i in table_rows[i]]
    length = len(df)
    df.loc[length] = individual_row
df
```
