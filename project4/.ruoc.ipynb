```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import datetime as dt

url = "https://coinmarketcap.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
table = soup.find("table")

head = table.find("thead").find("tr")
headers = [header.text for header in head]

body = table.find("tbody")
rows = body.find_all("tr")
BTC = [column.text for column in rows[0]]
ETH = [column.text for column in rows[1]]
BNB = [column.text for column in rows[3]]

fix_name = [name.text for name in body.find_all("p", class_="sc-71024e3e-0 ehyBa-d")]
fix_name = [fix_name[0], fix_name[1], fix_name[3]]

fix_market_cap = [
    market_cap.text
    for market_cap in body.find_all("span", class_="sc-11478e5d-1 jfwGHx")
]
fix_market_cap = [fix_market_cap[0], fix_market_cap[1], fix_market_cap[3]]

fix_volume = [
    volume.text
    for volume in body.find_all("p", class_="sc-71024e3e-0 bbHOdE font_weight_500")
]
fix_volume = [fix_volume[0], fix_volume[1], fix_volume[3]]

current_time = dt.datetime.now().strftime("%Y-%m-%d %H:%M:%S:%f")

df = pd.DataFrame(columns=headers[2:9])
df.loc[0] = BTC[2:9]
df.loc[1] = ETH[2:9]
df.loc[2] = BNB[2:9]

df["Name"] = fix_name
df["Market Cap"] = fix_market_cap
df["Volume(24h)"] = fix_volume
df["TimeStamp"] = current_time

df
```
